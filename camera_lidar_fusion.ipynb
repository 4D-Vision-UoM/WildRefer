```python
import numpy as np

# Load point cloud
pc = np.load('src/STRefer/points_rgbd/10/1617681350.912022272.npy')


# Separate geometry and color
xyz = pc[:, :3]        # 3D coordinates
rgb = pc[:, 3:6]       # RGB values (0-255)
rgb_norm = rgb / 255.0 # Normalized RGB [0, 1]
```

```python
# Visualize the point cloud with color information
import open3d as o3d

# Create a point cloud with color information
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(xyz)
pcd.colors = o3d.utility.Vector3dVector(rgb_norm)

# Visualize the point cloud
o3d.visualization.draw_plotly([pcd])
```

```python
# Save this pointcloud as a .ply file
o3d.io.write_point_cloud("sample_pointcloud.ply", pcd)
```
